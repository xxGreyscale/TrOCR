```python
from data.loader.custom_loader import CustomLoader

# Load the data
path = ["..\\datasets\\handwritten\\augmented_v2\\image_labels_dataset.csv"]
handwritten_df = CustomLoader(path)
handwritten_df.generate_dataframe()
# Shuffle the DataFrame
df = handwritten_df.get_dataframe()
shuffled_df = df.sample(frac=1).reset_index(drop=True)

# Update the DataFrame in the loader
handwritten_df.dataframe = shuffled_df
```
